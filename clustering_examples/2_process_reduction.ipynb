```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
plt.rc('font', size=10)
```

```python
palette_values2 =  ['#C17ADC', '#005D82']
colors2 = sns.color_palette(palette_values2)
colors2
```

```python
df_data = pd.read_csv("estimated_properties.csv")
df_data.head()
```

```python
df_data.columns
```

```python
fig = plt.figure(figsize=(12,12))
gs = GridSpec(3, 3, figure=fig)

ax_data1 = fig.add_subplot(gs[0, 0])
sns.histplot(data=df_data, x="MW", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[0, 1])
sns.histplot(data=df_data, x="isoelectric_point", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[0, 2])
sns.histplot(data=df_data, x="aromaticity", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[1, 0])
sns.histplot(data=df_data, x="aliphatic_index", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[1, 1])
sns.histplot(data=df_data, x="boman_index", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[1, 2])
sns.histplot(data=df_data, x="charge", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[2, 0])
sns.histplot(data=df_data, x="charge_density", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[2, 1])
sns.histplot(data=df_data, x="hydrophobic_ratio", hue="Activity", palette=colors2, fill=False)

ax_data1 = fig.add_subplot(gs[2, 2])
sns.histplot(data=df_data, x="instability_index", hue="Activity", palette=colors2, fill=False)

sns.despine()

plt.tight_layout()

```
