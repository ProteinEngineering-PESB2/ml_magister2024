```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
```

```python
def check_outlier_by_column(column_values=None, df_data=None, name_column=None):
    q1 = np.quantile(column_values, .25)
    q3 = np.quantile(column_values, .75)

    iqr = q3-q1

    lower = q1 - 1.5*iqr
    higher = q3 + 1.5*iqr

    array_data = []

    number_outlier = 0
    number_not_outlier = 0

    for index in df_data.index:
        value = df_data[name_column][index]

        if value>higher or value< lower:
            array_data.append(1)
            number_outlier+=1
        else:
            array_data.append(0)
            number_not_outlier+=1
    
    return array_data, number_outlier, number_not_outlier
```

```python
df_data = pd.read_csv("2_datos_no_nulos.csv")
```

```python
df_data
```

```python
df_data_outliers = pd.DataFrame()
```

```python
matrix_data = []

for column in df_data.columns:
    if column != "sex":
        df_data_outliers[column], number_outlier, number_not_outlier = check_outlier_by_column(
            df_data=df_data, 
            column_values=df_data[column].values, 
            name_column=column)
        
        row = [column, number_outlier, number_not_outlier]
        matrix_data.append(row)

df_summary_outliers = pd.DataFrame(data=matrix_data, columns=["column_name", "number_outlier", "number_not_outlier"])
```

```python
df_data_outliers["outlier_score"] = df_data_outliers.sum(axis=1, numeric_only=True)/(len(df_data.columns)-1)
df_data_outliers
```

```python
df_data_outliers["is_outlier"] = (df_data_outliers["outlier_score"]>0).astype(int)
df_data_outliers["is_outlier"].value_counts()
```

```python
clf = IsolationForest(
    random_state = 42
)

clf.fit(df_data)

df_data_outliers["is_outlier_if"] = clf.predict(df_data)
```
