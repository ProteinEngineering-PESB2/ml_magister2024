```python
import pandas as pd
```

```python
df_data = pd.read_csv("2_datos_no_nulos.csv")
df_data
```

```python
df_data = df_data.drop(columns=["height mt"])
df_data
```

```python
df_data["sex"].unique()
```

```python
matrix_data = []

for index in df_data.index:
    value = df_data["sex"][index]

    row = [0,0]

    if value == 1:
        row[0] = 1
    else:
        row[1] = 1
    
    matrix_data.append(row)

df_one_hot = pd.DataFrame(data=matrix_data, columns=["sex_1", "sex_2"])
df_one_hot
```

```python
df_data = pd.concat([df_data, df_one_hot], axis=1)
df_data
```

```python
df_data.drop(columns=["sex"], inplace=True)
```

```python
df_data.to_csv("4_dataset_procesed.csv", index=False)
```
