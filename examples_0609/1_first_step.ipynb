## Etapa 2: Analisis preliminar del dataset

```python
df_data.shape
```

```python
df_data = df_data.drop(columns=['Unnamed: 0', 'Unnamed: 1', 'redcap_repeat_instrument'])
df_data.head(5)
```

```python
df_data.dtypes
```

```python
df_data.describe()
```

```python
df_data["t2m_pre_rmssd"] = df_data["t2m_pre_rmssd"].astype(float)
```

```python
values_rmssd = [str(value).replace(" ", "").replace(",", ".") for value in df_data["t2m_pre_rmssd"].values]
df_data["t2m_pre_rmssd"] = values_rmssd
df_data["t2m_pre_rmssd"] = df_data["t2m_pre_rmssd"].astype(float)
```

```python
df_data.columns
```

```python
df_data["check_altura"] = ((df_data["height cm"]/100) == df_data["height mt"]).astype(int)
```

```python
df_data["check_altura"].value_counts()
```

```python
df_data[df_data["check_altura"] == 0]
```

```python
df_data["our_imc"] = (df_data["weight kg"]/ (df_data["height mt"] * df_data["height mt"]))
df_data["check_imc"] = ((df_data["weight kg"]/ (df_data["height mt"] * df_data["height mt"])) == df_data["imc"]).astype(int)
df_data["check_imc"].value_counts()
```

```python
df_data[df_data["check_imc"] == 0]
```

```python
df_out_imc = df_data[df_data["check_imc"] == 0]
df_data_filter_imc = df_data[df_data["check_imc"] != 0]
```

```python
df_out_imc.shape
```

```python
df_data_filter_imc.shape
```

```python
df_data_filter_imc["our_imc"] = (df_data_filter_imc["weight kg"]/ ((df_data_filter_imc["height cm"]/100) * (df_data_filter_imc["height cm"]/100)))
df_data_filter_imc["our_imc"].describe()
```

```python
df_data_filter_imc["our_imc"] = df_data_filter_imc["our_imc"].round(decimals=4)
```

```python
df_data_filter_imc[df_data_filter_imc["our_imc"] >= 57]
```

```python
df_data_filter_imc.iloc[393]
```

```python
df_data_filter_imc_2 = df_data_filter_imc[df_data_filter_imc["our_imc"]<57]
df_data_filter_imc_2.shape
```
